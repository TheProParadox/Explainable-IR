```python
def overlap_percentage(query, generated):
    query_words = set(query.lower().split())
    generated_words = set(generated.lower().split())
    overlap_count = len(query_words.intersection(generated_words))
    if len(query_words) == 0:
        return 0
    return (overlap_count / len(query_words)) * 100

df["Overlap"] = df.apply(lambda row: overlap_percentage(row["query"], row["CF_generatedd"]), axis=1)

# Display the first few rows to verify

df["Overlap"].mean()
```

```python
df['Num_new_words'] = df['New_words'].apply(lambda row: len(str(row).split()))

df['Num_new_words'].mean()
```
